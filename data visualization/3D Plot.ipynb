Here we use *iris* dataset to plot a 3D scatterplot.

**References**

http://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html#sphx-glr-auto-examples-datasets-plot-iris-dataset-py

http://matplotlib.org/mpl_toolkits/mplot3d/api.html

```python
%matplotlib inline

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :4]
Y = iris.target


# 1st plot
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(X[:, 1], X[:, 2], X[:, 3], c=Y,
           cmap=plt.cm.Paired)

ax.set_title("3D Test - 1")
ax.set_xlabel("1st feature")
# ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd feature")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd feature")
ax.w_zaxis.set_ticklabels([])


# 2nd plot
fig = plt.figure(2, figsize=(8, 6))
ax = Axes3D(fig, elev=0, azim=0)
ax.scatter(X[:, 1], X[:, 2], X[:, 3], c=Y)

ax.set_title("3D Test - 2")
ax.set_xlabel("1st feature")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd feature")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd feature")
# ax.w_zaxis.set_ticklabels([])


# 3rd plot
fig = plt.figure(3, figsize=(8, 6))
ax = Axes3D(fig, elev=90, azim=0)
ax.scatter(X[:, 1], X[:, 2], X[:, 3], c=Y)

ax.set_title("3D Test - 3")
ax.set_xlabel("1st feature")
# ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd feature")
# ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd feature")
ax.w_zaxis.set_ticklabels([])
```
